```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
df=titanic[['PassengerId','Survived','Pclass','Sex','Age','Fare','Embarked']]
df
```

```python
df.isna().sum()
```

```python
df[df['Embarked'].isna()]
```

```python
df['Embarked'].value_counts()
```

```python
#df['Embarked'].fillna(value='X')
#df=df[df['Embarked'].notna()]
df=df.dropna(subset=['Embarked'])
df.isna().sum()
```

```python
df.groupby('Pclass')['Age'].median()
```

```python
df['impAge']=df[['Pclass','Age']].groupby('Pclass').transform(lambda x: x.fillna(x.median()))['Age']
df.head()
df.isna().sum()
```

```python
df.pivot_table(values='PassengerId',index='Pclass',columns='Sex',aggfunc='count',margins=True)
```

```python
df.pivot_table(values='Survived',index='Pclass',columns='Sex',aggfunc='mean',margins=True)
```

```python
df['impAge'].describe()
```

```python
def getAgeGroup(x):
    if x <= 22:
        return'0-[,22]'
    elif 22< x <= 26:
        return '1-(22,26]'
    elif x>26 and x <=36:
        return '2-(26-36]'
    else:
        return '3-(36,]'
    
        
```

```python
df['agegroup']=df['impAge'].apply(getAgeGroup)
df.head()
```

```python
df.pivot_table(values='PassengerId',index='agegroup',columns='Sex',aggfunc='count',margins=True)
```

```python
df.pivot_table(values='Survived',index='agegroup',columns='Sex',aggfunc='mean',margins=True)
```
